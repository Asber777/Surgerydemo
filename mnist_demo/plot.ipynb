```python
import numpy as np
fs_acc = '/home/workspace/util/surgery/mnist_demo/0to7_fs/acces_0to7_fs'
su_acc = '/home/workspace/util/surgery/mnist_demo/0to7_surgery/acces_0to7_surgery'
```

```python
fs = np.loadtxt(fs_acc)
su=np.loadtxt(su_acc)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# make data
# x = np.linspace(0, 2000, 2000)
x = np.arange(0, 900, 50)
# plot
fig, ax = plt.subplots()
w = 1
ax.plot(x, fs, linewidth=w, label='from scratch')
ax.plot(x, su, linewidth=w, label='Surgery')
plt.xlabel("epoch")
plt.ylabel("test acc")
plt.legend()
plt.show()
```
